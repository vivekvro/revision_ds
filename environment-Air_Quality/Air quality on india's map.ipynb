```python
import pandas as pd
import json 
import numpy as np
import plotly.express as px
```

```python
india_states = json.load(open("Geojson files/states_india.geojson",'r'))
```

```python
df = pd.read_csv("Air-Quality-Maps-2023-Averaged.csv")
```

```python
state_id_map={}
```

```python
for feature in india_states['features']:
    feature["id"]=feature['properties']['state_code']
    state_id_map[feature['properties']['st_nm']] = feature['id']
```

```python
df['id'] = df['States/Ut'].apply(lambda X: state_id_map[X])
```

```python
fig = px.choropleth(df,locations='id',geojson=india_states,color='NO2 (in μg/m³)')
fig.update_geos(fitbounds="locations", visible=False)
fig.show()
```
