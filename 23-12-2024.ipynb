```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
url = 'https://raw.githubusercontent.com/Nirmit29/Student-Result-Analysis-DataSet-from-KAGGLE-/refs/heads/main/Expanded_data_with_more_features.csv'
df = pd.read_csv(url)
```

```python
df
```

```python
for i in df.select_dtypes(object).columns:
    print(df[i].value_counts())
```

```python
newdf = df['TestPrep'].replace(['none','completed'],[0,1])
```

```python
newdf = pd.DataFrame(newdf)
newdf[['MathScore','ReadingScore','WritingScore']] = df[['MathScore','ReadingScore','WritingScore']]
```

```python
newdf
```

```python
newdf= newdf.dropna(axis=0)
```

```python
newdf['TestPrep'] = newdf['TestPrep'].astype('int64')
```

```python
X = newdf[['MathScore','ReadingScore','WritingScore']].head(100)
y = newdf['TestPrep'].head(100)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=97)
lrg = LogisticRegression()
lrg.fit(X_train,y_train)
best_score = lrg.score(X_test,y_test)
best_score
```

```python
rdm =0
for i in range(1,200):
    X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=i)
    lrg.fit(X_train,y_train)
    score = lrg.score(X_test,y_test)
    if best_score < score:
        best_score=score
        rdm=i
    
rdm,best_score
```

```python
y_pred=lrg.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
report = classification_report(y_test, y_pred)
print(report)
```

```python
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", cm)
```
